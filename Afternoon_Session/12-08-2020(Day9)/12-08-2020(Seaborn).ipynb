```python
import seaborn as sns
import matplotlib.pyplot as plt
```

- Strip plot

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.set_style("whitegrid")
sns.stripplot(x="total_bill",data=tips)
```

```python
sns.stripplot(x="day",y="tip",data=tips)
```

```python
help(sns.stripplot)
```

```python
sns.stripplot(x="day",y="tip",data=tips,hue="size")
```

```python
sns.stripplot(x="day",y="tip",data=tips,hue="sex")
sns.swarmplot(x="day",y="tip",data=tips,hue="sex")
sns.barplot(x="day",y="tip",data=tips,hue="sex")
```

- Regression Plot
- lmplot()
- it creates a regression line between 2 parameters and helps to visualize their linear realationship

```python
import seaborn as sns
sns.lmplot(x="total_bill",y="tip",data=tips)
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",markers=["o","v"])
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",markers=["o","v"],palette="BuPu_r",
           scatter_kws={"s":100})
```

```python
help(sns.lmplot)
```

```python
#Displaying multiple plots
sns.lmplot(x="total_bill",y="tip",data=tips,col="sex",row="time",hue="smoker")
```

```python
#Displaying multiple plots
sns.lmplot(x="total_bill",y="tip",data=tips,col="sex",row="time",hue="smoker",
           height=6,palette="coolwarm")
```

```python
#bin represents how many falls in data
sns.distplot(tips["total_bill"],bins=20)
```
