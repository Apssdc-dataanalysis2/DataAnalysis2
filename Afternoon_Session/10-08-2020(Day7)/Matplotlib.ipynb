## Matplotlib
- Matplotlib is one of the library in python used for 2D plotting in Visualization
- Pyplot
    - Pyplot is a collection of functions. Each function can create a plotting area in figure, plot lines, plot with labels etc.
    
### Types of plotting in matplot
- Lines
- Bar Chart
- Scatter plot
- Histogram
- Box Plot
- Pie chart

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Line Potting
- plt.plot()

```python
x_axes = [1,2,3,4]
y_axes = [1,4,6,8]
plt.plot(x_axes,y_axes,'g')
plt.show()
```

```python
x_axes = [1,2,3,4]
y_axes = [1,4,6,8]
plt.plot(x_axes,y_axes,'g')
plt.xlabel('X Axes')
plt.ylabel('Y Axes')
plt.title('Line')
plt.show()
```

```python
x = np.linspace(0,2,10)
x
```

```python
plt.plot(x, x, label = 'Line')
plt.plot(x, x**2, label = 'Square')
plt.legend(loc = 'best')
```

```python
help(plt.legend)
```

```python
plt.xticks([2,4,6,8])
plt.yticks([2,4,6,8])
```

## Bar Chart
- plt.bar()

```python
plt.bar([1,2,3,4,5],[2,6,4,8,10],width = 0.33,color = 'k')
```

```python
x = np.arange(4)
x
```

```python
plt.bar(x,[89,90,55,60],color = 'b', width = 0.33)
plt.bar(x + 0.33,[85,88,96,48],color = 'y', width = 0.33)
```

## Histogram
- Histogram is similar to bar chart. Used for frequency distribution.
- plt.hist()

```python
a = [0,0,0.2,0.3,2,2.2,2.5,4,4.1,4.4,5.1]
b = 5
plt.hist(a,b,rwidth = 0.4)
```

```python
a2 = [22,22,23,23.4,25,25,25.1,26,26.0,28,28.3,28.4,28.5,28.6,30,30,30.1,30.2]
bin2 = [21,23,24,25,26,27,28,29,31]
plt.hist(a2,bin2,rwidth = 0.4)
```

## Pie Chart
- plt.pie()

```python
s = ['Data Analysis','Machine Learning','Python']
marks = [80,90,75]
plt.pie(marks,colors = ['b','g','y'],labels = s,autopct = '%1.2f%%',startangle = 90)
plt.show()
```
