#### Countplot

```python
sns.set_style("dark")
sns.countplot(x="sex",data=df)
```

```python
tips = sns.load_dataset("tips")
sns.set_style("whitegrid")
sns.countplot(x="sex",data=tips)
```

```python
plt.figure(figsize=(12,3))
sns.countplot(x="sex",data=tips)
```

```python
# Scale and context
sns.set_context("poster",font_scale=2)# poster,paper,notebook,talk
plt.figure(figsize=(20,4))
sns.countplot(x="sex",data=tips)
```

#### Swarmplot

```python
sns.set_context("notebook",font_scale=1)
sns.swarmplot(x=tips["total_bill"])
```

```python
tips.columns
```

```python
sns.set_context("notebook",font_scale=1)
sns.swarmplot(x="day",y="total_bill",data=tips)
```

```python
sns.swarmplot(x="day",y="total_bill",hue="size",data=tips)
```

```python
sns.swarmplot(x="day",y="total_bill",hue="sex",data=tips)
```

```python
tips.columns
```

```python
sns.swarmplot(x="time",y="tip",data=tips)
```

```python
sns.swarmplot(x="time",y="tip",data=tips,order=["Dinner","Lunch"])
```
