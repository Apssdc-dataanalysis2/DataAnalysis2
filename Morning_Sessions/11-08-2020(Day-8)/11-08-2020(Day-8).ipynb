### Histogram:

* Frequency distrubution of the given numerical data.


```python
help(plt.hist)
```

```python
a = [0,1,1.5,0.5,2,3,3.45,5.9,1.4,9,6,15]

plt.hist(a,bins = 5,color = 'm',width = 1)
plt.show()
```

```python
b = [0,5,10,15,4,3,2,2.5,6,14]
plt.hist(b,rwidth = 0.5,histtype = "stepfilled",color = 'g')
plt.savefig("sample1.jpg")   ### saving the image
plt.show()
```

### Box plot:

* Box plot is also named as Whisker plot.
* The main terms used in box plot are minimum,maximum,first quartile,median,third quartile
* First


```python
data = np.random.normal(100,20,200)
data
```

```python
plt.boxplot(data)
```

```python
a = np.random.randint(50,100,10)
b = np.random.randint(40,90,10)
c = np.random.randint(30,100,10)
plt.boxplot([a,b,c],labels = ["box1","box2","box3"])
plt.show()
```

```python
help(plt.boxplot)
```

###  Area Plot:


```python
days = [1,2,3,4]
working = [8,11,9,10]
sleeping = [8,10,11,12]

plt.stackplot(days,working,sleeping,colors = ['r','g'],labels = ['work','slept'])
plt.legend()
plt.show()
```

### Pie plot:

```python
subjects = ["Python","C","Java"]
Marks = [25,50,25]

plt.pie(Marks,labels = subjects,colors = ['r','g','m'],radius = 1,explode = [1,1,1])
plt.show()
```
