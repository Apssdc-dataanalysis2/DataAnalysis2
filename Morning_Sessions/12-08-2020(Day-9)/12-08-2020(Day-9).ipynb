```python
import seaborn as sns
iris = sns.load_dataset("iris")  ### accessing default datasets available in seaborn
iris.columns
```

```python
sns.jointplot(x = "sepal_length",y = "sepal_width",data = iris,kind = "reg")
```

```python
sns.get_dataset_names()   ### all available datasets.
```

```python
sns.jointplot(x = "petal_length",y = "petal_width",data = iris,kind = "reg")
```

```python
sns.jointplot(x = "petal_length",y = "petal_width",data = iris)
```

```python
sns.jointplot(x = "petal_length",y = "petal_width",data = iris,kind = "kde")
```

```python
sns.jointplot(x = "petal_length",y = "petal_width",data = iris,color = "k").plot_joint(sns.kdeplot,n_levels = 10)
```

```python
import numpy as np
x,y = np.random.randn(2,300)
g = sns.jointplot(x,y,kind = "hex").set_axis_labels('x','y')
```

```python
sns.jointplot("petal_length","petal_width",data = iris,kind = "hex")
```

```python
sns.jointplot("sepal_length","petal_width",data = iris,kind = "hex")
```

```python
### pairplot:

sns.set_style("ticks")
sns.pairplot(iris,hue = "species",diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()
```

```python
sns.pairplot(iris,hue = "species",markers = ["o","s","d"])
plt.show()
```

```python
### stripplot:

sns.stripplot(x = "species",y = "sepal_length",data = iris)
plt.show()
```

```python
### heatmap:

a = np.random.randn(10,20)
g = sns.heatmap(a)
plt.show()
```

```python
sns.heatmap(a,vmin = 0,vmax = -5)
```

```python
plt.figure(figsize = (10,5))
sns.heatmap(a,annot = True)
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.heatmap(a,annot = True,linewidth = 0.5)
```

```python
sns.heatmap(a,cmap = "YlGnBu")
```
