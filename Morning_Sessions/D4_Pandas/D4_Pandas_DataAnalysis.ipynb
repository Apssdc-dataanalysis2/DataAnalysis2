## creating series

````python
# Create pandas Series using List
se1 = pd.Series([234,56,768,89,98])
se1
````

````python
# Series using tuple
se2 = pd.Series((234,56,768,89,98),index = ["d","df","t",56,56.76])
se2
````

````python
# Series using Dictionary
dic = { "Name" : ["chaitanya","siva Kumar","Naresh","Geetha","siva shankar"]
       }
type(pd.Series(dic))
print(pd.Series(dic))
````

````python
# Series using Numpy array
import numpy as np
num = np.arange(1,11)
se4 = pd.Series(num**2,index = np.arange(1,11))
se4
````

## Create DataFrame

````python
# Df using List
df1 = pd.DataFrame([[1,2,3],["a","t","v"],[12.23,34.5,6.5]])
df1
````

````python
df1.columns = ["col1","col2","col3"]
df1
````

````python
df1.index = [10,20,30]
df1
````

````python
# Df using Dictionary  # keys acts as label/ column/ feature names
di = { "Name" : ["chaitanya","siva Kumar","Naresh","Geetha","siva shankar"],
        "Age" : [23,24,16,26,34],
       "Gender" : ['m',"m","m","f","m"]
     }
df2 = pd.DataFrame(di,index = np.arange(1,6))
df2
````

````python
# df using using series
di = { "Name" : pd.Series(["chaitanya","siva Kumar","Naresh","Geetha","siva shankar"]),
        "Age" : pd.Series([23,24,16,26,34]),
       "Gender" : pd.Series(['m',"m","m","f","m"])
     }
df2 = pd.DataFrame(di,index = np.arange(0,5))
df2
````

````python

di = { "Name" : pd.Series(["chaitanya","siva Kumar","Naresh","Geetha","siva shankar"],index = [1,2,3,4,5]),
        "Age" : pd.Series([23,24,16,26],index=[1,2,3,4]),
       "Gender" : pd.Series(['m',"m","f","m"],index=[1,3,4,5])
     }
df3 = pd.DataFrame(di,index = np.arange(1,6))
df3
````

## Combining  

````python
pd.concat([df2,df3],axis = 0) # row concat axis 0
````

````python
pd.concat([df2,df3],axis = 1)
````

````python
df2.append(df3)
````

````python
df2.append(df3)
````

````python
pd.merge(df2,df3) # rows that contain the exact same values for the common column labels will be merged
````
