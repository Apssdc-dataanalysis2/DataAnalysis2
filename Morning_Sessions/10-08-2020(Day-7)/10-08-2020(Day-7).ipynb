## Data Representation with graphical formats:


For grahical representation we mainly use 2 libraries:
* Matplotlib
* Seaborn


### Matplotlib:


* Matplotlib is a python library used to represent the data in graphical format.
* 2dgraphs can be drawn by using matplotlib.
* Pyplot Module is used to reprensent our data in graphical format.
* It is a cross-platform for data visualization library for making 2d plots from the given data.

### Types of Plotting:

* Line plot
* Bar Graph
* Scatter plot
* Area Plot
* pie charts
* Histogram
* Box plot




```python
#### Importing matplotlib library:

import matplotlib.pyplot as plt
```

### Line plot:



```python
x = [1,2,3,4]
y = [1,2,3,4]

plt.plot(x,y)  ### plotting the graph for variables x and y
plt.show()
```

```python
plt.plot(x,y,'g')  ### color formating of the line.
plt.show()
```

```python
plt.plot(x,y,'g')
plt.title("Sample plot")  ### title of the graph
plt.xlabel("X-axis")     ### name of the x-axis
plt.ylabel("y-axis")     ### name of the y-axis
plt.show()              ### showing the graph
```

```python
import numpy as np
x = np.linspace(0,4,50)
x
```

```python
plt.plot(x,x,color = 'r',label = "sample1")
plt.plot(x,x**2,color = 'b',label = "sample2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("line plot")
plt.legend()
plt.show()
```

###  subplots:

plt.subplot(nrows,ncols,index)

###   Scatter plot:



```python
x = np.arange(1,50,5)

y = np.random.randint(1,50,size = len(x))

print(x,y)
```

```python
plt.scatter(x,y)
plt.text(6,34,'(6,34)')
plt.show()
```

```python
help(plt.scatter)
```

```python
plt.scatter(x,y,color = "m",marker = "*")
```

### Bar graph:



```python
plt.bar(x,y,width = 1.5)
plt.show()
```
